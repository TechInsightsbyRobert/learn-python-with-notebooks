## üîÇ What Is a Loop?

A loop is a way to tell python:
> ‚ÄúRun this block of code more thaan once.‚Äù  

Instead of copying and pasting the same code 10 times, you can use a loop to do the repetition for you.

### ü§î Why This Matters
Loops are everywhere in programming:
- Repeating actions (printing, calculating, checking)
- working with lists of data
- Counting, scoring, or tracking progress
- Automating boring or repetitve tasks

If you want to write efficient and powerful code, loops are essential.

## üîÇ For Loops
### üß† What Is a `for` Loop?
A for loop is used when you know how many times you want something to repeat

You are telling Pyhton:
> ‚ÄúFor each value in a sequence, do this.‚Äù  

### üí° Why Beginners Use `for` Loops
- They are predictable and easy to read
- Greate for counting or repeating steps
- Very common in beginner projects and data work

## üî¢ Loop Counters

## üí° Why Loop Counters Matter

- Useful for numbering items
- Common in gmaes, scores, and repetition logic

## üîÑ While Loops

A `while loop` repeats code as long as a condition is true.

Instead of saying *how maany times*, you say"
> "Keep Going while this is true."

### üí° Why Beginners Use `while` Loops
- Useful when you don't know how many times something should repeat
- Common in games, input checks, and simulations
- Gives more control, but requires more care
---
## ‚úÖ Task 4: Your First `while` Loop
```python
count = 0

while count < 3:
    print("Counting:", count)
    count += 1

```
### üß† What‚Äôs happening?
- `Count` starts at 0
- The loop checks the condition: `count < 3`
- Code runs while the condition is true
- `count += 1` prevents an infinite loop
