## üß† What Is a Function?

### Explanation
A **function** is a named block of code that performs a specific task.

Think of a function like a **machine**:
- You give it input
- It does some work
- It may give something back

Instead of writing the same code over and over, you write it **once** and reuse it.

### Why This Matters
Functions help you:
- Avoid repeating code
- Make programs easier to read
- Break big problems into smaller pieces

Beginners use functions whenever a task needs to be done **more than once**.

### üß† Why This Matters:

- With parameters, they become flexible and powerful.
