## üß© Lists: Storing Multiple Values

### Explanation
A **list** is a collection of items stored in a specific order.  
Each item has a **position**, called an **index**, starting at 0.

Example: a grocery list, a to-do list, or a collection of numbers.

### Why This Matters
Lists allow you to manage **related items together** instead of creating separate variables for each item. This makes your code cleaner and easier to work with.

---

### üß† What‚Äôs happening?
- `fruits stores three items

- You access items using square brackets `[  ]`

## ‚ûï Adding and Removing Items from a List

Lists are flexible ‚Äî you can **add**, **remove**, or **change** items after creation.
- `appened()` -> adds an item to the end
- `insert()` -> adds an item at a specific position
- `remove()` -> removes an item by value
- `pop()` -> removes an item by index
  
Being able to modify lists is essential when working with **dyanmice data**, like user input or changing datasets.

---
## ‚úÖ Task 2: Modifying a List
```python
fruits = ["apple", "banana", "cherry"]

fruits.append("orange")
fruits.remove("banana")
fruits.insert(1, "kiwi")

print(fruits)

```
### üß† What‚Äôs happening?
- `append()` added "orange" to the end
- `remove()` deleted "banana"
- `insert()` placed "kiwi" at index 1

## üîë Dictionaries: Key-Value Pairs

A **dictionary** is a collection of **key-value pairs**. Each key is unique and maps to a value.

> Think of a phonebook - the key is the person's name, and the vallue is their phone number.

Dictionaries are useful when you want to **associate information**, like storing user data, grades, or product prices.

---
## ‚úÖ Task 3: Creating and Accessing Dictionaries
```python
person = {
    "name": "John",
    "age": 20,
    "city": "Huston"
}

print(person["name"])
print(person["age"])

```
### üß† What‚Äôs happening?
- Each key (`"name"`, `"age"`) points to a value (`"Robert"`, `20`)
- You have access values by their keys
- Keys can be strings, numbers, or even tuples, but most beginners use strings

## üîÑ Modifying Dictionaries

Dictionaries are dynamic. You can update, add, or delete key-value pairs.
- Update a value: `dict[key] = new_value`
- Add a pair: `dict[new_key] = value`
- Delete a pair: `del dict[key]`
  
Real-world data changes all the time. Being able to modify dictionaries makes programs **flexible** and **practical**.
