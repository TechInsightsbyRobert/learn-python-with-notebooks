## üîÅ Variable Reassignment

Python variables are not locked to one value.

When you reassign a variable, Python:

2. Keeps the same variable name
3. Uses the new value from that point on

## üß© What Are Data Types?
A data type tells Python:
- What kind of value something is
- What operations are allowed on it
Different data types behave differently.

### üî§ Common Beginner Data Types
| Type    | Name                   | Example       |
| ------- | ---------------------- | ------------- |
| `int`   | Integer (whole number) | `5`, `-3`     |
| `float` | Decimal number         | `3.14`, `0.5` |
| `str`   | String (text)          | `"Hello"`     |

---
## üß™ Understanding Data Types with type()

Python automatically assigns a data type to every value.

The type() function lets us ask Python what type something is.

---
## ‚úÖ Task 3: Checking Data Types 
```python
x = 100
y = 3.14
z = "Python is fun!"

print(type(x)) # Prints: <class 'int'>
print(type(y)) # Prints: <class 'float'>
print(type(z)) # Prints: <class 'str'>

```
### üß† Common Data Types
- `int` -> whole numbers
- `float` -> decimals
- `str` -> text (strings)

## üîÑ Type Casting (Changing Data Types)

Sometimes you need to convert data from one type to another.

Type casting means converting one data type into another.

We do this when:
- We get user input
- We want to perform math
- Python refuses an operation

Common conversions:
- `int()` ‚Üí convert to whole number
- `float()` ‚Üí convert to decimal
- `str()` ‚Üí convert to text 

---
## ‚úÖ Task 5: Type Casting with Numbers
```python

age = int(age)

print("Next year you will be", age + 1)

```
### üß† Why do we do this?
- `input()` gives us a string
- Math only works with numbers
- We convert the string to a number using `int()`
