## ğŸ” `if`, `elif`, and `else`

Sometimes you have multiple choices, not just yes or no.

When you reassign a variable, Python:
- `if` checks the first condition
- `elif` (else if) checks another condition if the first is false
- `else` runs if all previous conditions are false

This allows python to handle multiple scenarios, whihc is very common in real-world programs
